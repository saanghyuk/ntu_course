```python
d = {'ISO3': all_iso3_list,
 'Country' : coco.convert(names=all_iso3_list, to="name_short"), 
    'MPPL' : [12.750179468772433]*len(all_iso3_list)}

append_df = pd.DataFrame(data=d)
append_df.head()
```

```python
concentration.head()
```

```python
concentration_before_concat = concentration[["ISO3", "Country","MPPL"]]
```

```python
concentration = pd.concat([concentration_before_concat, append_df])
concentration = concentration.reset_index()
concentration = concentration.drop("index", axis=1)
```

## Merge all data to make one integrated CSV file

### Merge concentration with production

```python
concentration.head()
```

```python
production_groupby["ISO3"] = production_groupby.index
production_groupby.head()
```

```python
production_groupby.reset_index(drop=True, inplace=True)
production_groupby.head()
```

```python
result = pd.merge(concentration, production_groupby, how= 'left', on=["ISO3"])
```

```python
result.rename(columns={"[2019]": "Production_2019"}, inplace=True)
result.head()
```

### Merge trade

```python
import_df = trade_import.groupby(by=["ISO3_reporting_country"], dropna=False).sum()
import_df["ISO3"] = import_df.index
import_df.reset_index(drop=True, inplace=True)
import_df.head()
```

```python
import_df = import_df.iloc[:, 1:]
```

```python
import_df.rename(columns={"[2019]": "Imports"}, inplace=True)
import_df.head()
```

```python
result.head()
```

```python
result = pd.merge(result, import_df, how= 'left', on=["ISO3"])
```

```python
result.head()
```

```python
export_df = trade_export.groupby(by=["ISO3_reporting_country"], dropna=False).sum()
export_df["ISO3"] = export_df.index
export_df.reset_index(drop=True, inplace=True)
export_df.head()
```

```python
export_df.rename(columns={"[2019]": "Exports"}, inplace=True)
export_df.head()
```

```python
export_df = export_df.iloc[:, 1:]
export_df.head()
```

```python
results = pd.merge(result, export_df, how= 'left', on=["ISO3"])
results.head()
```

### Export as CSV

```python
results.to_csv("concat_table.csv")
```
