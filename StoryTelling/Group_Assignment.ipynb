```python
import pandas as pd
import numpy as np
import geocoder
```

### AdventureScientist.csv

```python
AdventureScientist = pd.read_csv("./Data/AdventureScientist.csv") 
AdventureScientist.head()
```

```python
as_lat = AdventureScientist["Latitude"]
as_long = AdventureScientist["Longitude"]
```

```python
as_country = []
for lat, long in zip(as_lat, as_long):
    print(f"lat {lat}, long {long}")
    g = geocoder.osm([lat, long], method='reverse')
    try: 
        print(f"Country Name : {g.json['country']}")
        as_country.append(g.json['country']) 
    except:
        print("Not Found")
        as_country.append("Not Found") 
```

```python
AdventureScientist["Country"] = as_country
```

```python
AdventureScientist.to_csv('AdventureScientist.csv')
```
